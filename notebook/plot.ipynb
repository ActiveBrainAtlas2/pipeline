```python
from pathlib import Path

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import numpy as np
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

def plot(data_dir, moving_tag, warped_tag, pdf_tag,
        warped_title, fixed_title, diff_title):
    data_dir = Path(data_dir)
#     moving_image = np.load(data_dir / f'{moving_tag}.npy')
    warped_image = np.load(data_dir / f'{warped_tag}.npy')
    fixed_image = np.load(data_dir / 'fixed_image.npy')
    diff_image = warped_image - fixed_image
    
    with PdfPages(data_dir.parent / f'{pdf_tag}.pdf') as pdf:
        sz = fixed_image.shape[-1]
        for z in range(0, sz, 2):
            print(f'{z}/{sz}')
            px = 1 / plt.rcParams['figure.dpi']
            fig, ax = plt.subplots(1, 3, dpi=250, figsize=(8, 6))
            kwargs = {
                'aspect':'equal',
                'cmap': 'gray',
            }
#             ax[0].imshow(moving_image[:,:,z], **kwargs)
#             ax[0].set_title('moving')
            ax[0].imshow(warped_image[:,:,z], **kwargs)
            ax[0].set_title(warped_title)
            ax[0].set_axis_off()
            ax[1].imshow(fixed_image[:,:,z], **kwargs)
            ax[1].set_title(fixed_title)
            ax[1].set_axis_off()
            kwargs['cmap'] = 'coolwarm'
            kwargs['vmin'] = -1
            kwargs['vmax'] = 1
            ax[2].imshow(diff_image[:,:,z], **kwargs)
            ax[2].set_title(diff_title)
            ax[2].set_axis_off()
            fig.suptitle(f'z = {z}')
            fig.tight_layout()
            pdf.savefig(fig)
            plt.close()
```
